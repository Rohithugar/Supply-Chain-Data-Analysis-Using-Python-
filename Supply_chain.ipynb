```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
numerical_columns = data.select_dtypes(include=[np.number]).columns
for col in numerical_columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(data[col], bins=30, kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# Select only numeric columns for correlation analysis
numerical_columns = data.select_dtypes(include=[np.number])

# Compute the correlation matrix
correlation_matrix = numerical_columns.corr()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 7. Relationships Between Key Variables
# Revenue vs. Manufacturing Costs
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='Manufacturing costs', y='Revenue generated')
plt.title('Revenue vs. Manufacturing Costs')
plt.show()
```

```python
# Shipping Costs vs Shipping Times
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='Shipping times', y='Shipping costs', hue='Transportation modes')
plt.title('Shipping Costs vs. Shipping Times')
plt.show()
```

```python
# 8. Outlier Detection Using Boxplots
for col in numerical_columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=data[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
# 9. Category Counts and Visualizations
# Count of Product Types
plt.figure(figsize=(10, 5))
sns.countplot(data=data, x='Product type')
plt.title('Count of Product Types')
plt.xticks(rotation=45)
plt.show()
```

```python
# Supplier Count Distribution
plt.figure(figsize=(10, 5))
sns.countplot(data=data, y='Supplier name')
plt.title('Count of Products by Supplier')
plt.show()
```

```python
# 10. Revenue Generated by Product Type
revenue_by_product = data.groupby('Product type')['Revenue generated'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=revenue_by_product, x='Product type', y='Revenue generated', palette='viridis')
plt.title('Total Revenue by Product Type')
plt.xticks(rotation=45)
plt.show()
```

What is the total revenue generated by each product type?

```python
revenue_by_product_type = data.groupby('Product type')['Revenue generated'].sum().reset_index()
revenue_by_product_type
```

Which SKU has the highest revenue?

```python
top_sku = data.loc[data['Revenue generated'].idxmax()]
top_sku[['SKU', 'Revenue generated']]
```

```python
avg_defect_rate_by_product = data.groupby('Product type')['Defect rates'].mean().reset_index()
avg_defect_rate_by_product
```

Which supplier has the highest manufacturing cost?

```python
highest_cost_supplier = data.loc[data['Manufacturing costs'].idxmax()]
highest_cost_supplier[['Supplier name', 'Manufacturing costs']]
```

What are the most common shipping carriers used?

```python
carrier_counts = data['Shipping carriers'].value_counts().reset_index()
carrier_counts.columns = ['Shipping carriers', 'Count']
carrier_counts
```

Which transportation mode incurs the highest average cost?

```python
transportation_costs = data.groupby('Transportation modes')['Costs'].mean().reset_index()
transportation_costs.sort_values(by='Costs', ascending=False)
```

What is the relationship between defect rates and inspection results?

```python
inspection_defect_relation = data.groupby('Inspection results')['Defect rates'].mean().reset_index()
inspection_defect_relation
```

What is the correlation between lead times and stock levels?

```python
corr_lead_time_stock = data[['Lead times', 'Stock levels']].corr()
corr_lead_time_stock
```

Which product type has the highest defect rate?

```python
highest_defect_product = data.groupby('Product type')['Defect rates'].mean().reset_index()
highest_defect_product.sort_values(by='Defect rates', ascending=False).head(1)
```

 What are the average lead times for each supplier?

```python
avg_lead_time_by_supplier = data.groupby('Supplier name')['Lead time'].mean().reset_index()
avg_lead_time_by_supplier
```

Which product type has the highest availability?

```python
highest_availability = data.groupby('Product type')['Availability'].mean().reset_index()
highest_availability.sort_values(by='Availability', ascending=False)
```

 Which product has the highest production volume?

```python
top_production_sku = data.loc[data['Production volumes'].idxmax()]
top_production_sku[['SKU', 'Production volumes']]
```

Which routes are associated with the highest shipping costs?

```python
shipping_cost_by_route = data.groupby('Routes')['Shipping costs'].mean().reset_index()
shipping_cost_by_route.sort_values(by='Shipping costs', ascending=False)
```
